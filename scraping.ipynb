```python
import pandas as pd
from newsapi import NewsApiClient

newsapi = NewsApiClient(api_key='3fad2cdbf53c444da243516570c04962')

def get_stock_headlines(stock_symbol):
    try:
        headlines = newsapi.get_everything(
            q=stock_symbol,
            language='en',
            sort_by='publishedAt',
            page_size=100,
            exclude_domains='anotherdomain.com',  # Add domains to exclude
        )
        headlines_list = []
        for article in headlines['articles']:
            title = article['title']
            source = article['source']['name']
            url = article['url']
            if stock_symbol.lower() in title.lower():
                headlines_list.append({'Source': source, 'Title': title, 'URL': url})

        df = pd.DataFrame(headlines_list)

        print(df)

        df.to_csv(f'{stock_symbol}_headlines.csv', index=False)

    except Exception as e:
        print(f"An error occurred: {str(e)}")
get_stock_headlines('AAPL')
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import yfinance as yf

def get_company_name(ticker):
    try:
        stock_info = yf.Ticker(ticker)
        company_name = stock_info.info['longName']
        return company_name
    except:
        return None

def get_stock_news(ticker, num_results=100):
    company_name = get_company_name(ticker)

    if company_name:
        headlines = []
        links = []
        num_pages = (num_results // 10) + 1

        for page in range(num_pages):
            start_index = page * 10
            search_url = f'https://www.google.com/search?q={company_name}+Stock+News&start={start_index}'
            response = requests.get(search_url)
            if response.status_code == 200:
                soup = BeautifulSoup(response.text, 'html.parser')
                headlines.extend([headline.text for headline in soup.find_all('h3')])
                links.extend([a['href'] for a in soup.find_all('a', href=True) if 'url?q=' in a['href']])
        df = pd.DataFrame({'Headlines': headlines[:num_results], 'Links': links[:num_results]})
        df['Links'] = df['Links'].str.replace(r'/url\?q=', '')
        csv_filename = f'{company_name}_stock_news.csv'
        df.to_csv(csv_filename, index=False)
        print(f'Data saved to {csv_filename}')

        return df
    else:
        print(f"Unable to fetch company information for ticker {ticker}. Please check the ticker symbol.")
        return None

ticker = 'AAPL'
stock_news_df = get_stock_news(ticker, num_results=50)

if stock_news_df is not None:
    stock_news_df
```
